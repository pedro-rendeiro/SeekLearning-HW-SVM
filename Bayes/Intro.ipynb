## 1) Deduza o teorema de Bayes simplesmente usando a definição de probabilidade condicional.

A probabilidade condicional afirma que, dados dois eventos A e B, a probabilidade do evento A acontecer dado que um evento B já aconteceu é definida por $P(A/B) = {P(A,B) \over P(B)}$, em que P(A,B) é a probabilidade conjunta, ou seja, a probabilidade de A e B ocorrerem simultaneamente. <br>
Ao manipular essa expressão, pode-se obter $P(A,B) = P(B)P(A/B)$$^{[2]}$, ou seja, a probabilidade conjunta pode ser definida como a probabilidade de B ocorrer e a probabilidade de A ocorrer dado que B ocorreu.<br>
Ao considerar a "ordem" contrária dos eventos, ou seja, que A aconteceu primeiro, pode-se afirmar que $P(B/A) = {P(B,A) \over P(A)}$ e $P(B,A) = P(A)P(B/A)$. Como $P(A,B) = P(B,A)$, é possível substituir a probabilidade conjunta a partir da segunda expressão$^{[2]}$ e obter: $P(B)P(A/B) = P(A)P(B/A)$.<br>
O teorema de Bayes vem da manipulação dessa expressão, o qual afirma que $P(A/B) = {P(B/A)P(A) \over P(B)}$
